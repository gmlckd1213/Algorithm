```python
def coinChange(n):
    result = 0
    coins = [100, 50, 10, 5, 1]
    
    for coin in coins:
        result += n//coin
        n = n % coin
    return result

def main():
    n = int(input())
    print(coinChange(n))
if __name__ == "__main__":
    main()
```
