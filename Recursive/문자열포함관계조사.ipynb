```python
def check(A, B):
    if A in B:
        return "Yes"
    char = A
    if char[0] in B:
        char = char[1:]
    if char != '' and char == A:
        return "No"
    return check(char,B)
    
            

def main():
    A = input()
    B = input()
    print(check(A, B))

if __name__ == "__main__":
    main()
```
