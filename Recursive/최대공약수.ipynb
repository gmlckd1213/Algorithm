```python
def GCD(x, y):
    if x % y == 0:
        return y
    return GCD(y, x%y)

def main():
    data = [int(v) for v in input().split()]
    x = data[0]
    y = data[1]
    print(GCD(x,y))
if __name__ == "__main__":
    main()
```
