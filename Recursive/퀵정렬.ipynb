```python
def quickSort(array):

    
    if len(array) <= 1:
        return array 
    left, mid, right = [], [], []
    pivot = array[len(array)//2]
    for i in  array:
        if i < pivot:
            left.append(i)
        elif i>pivot:
            right.append(i)
        else:
            mid.append(i)
    return quickSort(left)+mid+quickSort(right)

def main():
    line = [int(x) for x in input("   (:10 2 3 4 5 6 9 7 8 1): ").split()]

    print(' :', *quickSort(line))

if __name__ == "__main__":
    main()



```
