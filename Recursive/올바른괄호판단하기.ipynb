```python
def check(m):
    if len(m) == 0:
        return "YES"
    elif len(m) == 1:
        return "NO"
    for i in range(len(m)):
        if m[i] == '(' and m[i+1] == ')':
            q = m[:i] + m[i+2:]
            return check(q)
    return "NO"


def main():
    gal = input()
    print(check(gal))

if __name__ == "__main__":
    main()
```
