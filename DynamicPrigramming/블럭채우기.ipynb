```python
def fillBox(n) :
    '''
    2 x n   2 x 1      1,000,000,007     .
    '''
    T = [1, 1]
    
    for i in range(2, n+1):
        T.append((T[i-1]+T[i-2])%1000000007)
    return T[n]
```

```python
def main():
  
    n = int(input())

    print(fillBox(n))

if __name__ == "__main__":
    main()
```
