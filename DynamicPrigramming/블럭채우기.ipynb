```python
def fillBox(n) :
    '''
    2 x n 의 상자를 2 x 1 의 블럭으로 채우는 경우의 수를 1,000,000,007로 나눈 나머지를 반환하는 함수를 작성하세요.
    '''
    T = [1, 1]
    
    for i in range(2, n+1):
        T.append((T[i-1]+T[i-2])%1000000007)
    return T[n]
```

```python
def main():
  
    n = int(input())

    print(fillBox(n))

if __name__ == "__main__":
    main()
```
