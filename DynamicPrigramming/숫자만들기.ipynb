```python
def makeNumber(n, m) :

    
    result = [1]
    
    for i in range(1, n+1):
        sum_temp = 0
        for j in range(1, min(i,m)+1):
            sum_temp += result[i-j]
        result.append(sum_temp%1000000007)
    return result[n]
```

```python
def main():
   
    firstLine = [int(x) for x in input().split()]

    n = firstLine[0]
    m = firstLine[1]

    print(makeNumber(n, m))

if __name__ == "__main__":
    main()
```
