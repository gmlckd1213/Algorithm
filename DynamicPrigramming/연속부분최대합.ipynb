```python
def getSubsum(data) :
   
    maxSum = []
    final_max = data[0]
    maxSum.append(data[0])
    
    for n in range(1,len(data)):
        maxSum.append(max(maxSum[n-1], 0) + data[n])
        if final_max < maxSum[n]:
            final_max = maxSum[n]
    return final_max
    
  
```

```python
import sys



def main():

    data = [int(x) for x in input().split()]

    print(getSubsum(data))

if __name__ == "__main__":
    main()
```
