```python
def eating(data) :
    '''
         list  ,          .
    '''
    days = len(data)
    d = [[0]*3 for i in range(days+1)]
    
    for i in range(1, days+1):
        for j in range(3):
            for k in range(3):
                if j == k :
                    continue
                d[i][j] = max(d[i][j], d[i-1][k])
            d[i][j] += data[i-1][j]
    return max(d[days][i] for i in range(3))
```

```python
def main():
    '''
       .
    '''

    n = int(input())

    data = []

    for i in range(n) :
        __line = [int(x) for x in input().split()]
        data.append(__line)

    print(eating(data))

if __name__ == "__main__":
    main()
```
